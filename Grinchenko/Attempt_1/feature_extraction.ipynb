```python
import pandas as pd
from tsfresh.feature_extraction import extract_features
from tsfresh.feature_extraction.settings import MinimalFCParameters
from tqdm import trange
```

MinimalFCParameters() - is one of the default sets of feature calculators, which is not so computationally consuming.

But in the future, we have to more carefully select extracting features

```python
for i in trange(1,9):
    data = pd.read_pickle(f"split_{i}.pkl")
    features = extract_features(timeseries_container=data, 
                            column_id="customer_ID", 
                            column_sort="S_2", 
                            default_fc_parameters=MinimalFCParameters(), 
                            n_jobs=16)
    del data
    features.to_pickle(f"feat_split_{i}.pkl")
    del features
```
